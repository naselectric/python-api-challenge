# Import Modules

```python
from citipy import citipy
import pandas as pd
import numpy as np
import json 
import requests
import matplotlib.pyplot as plt
import random
from datetime import datetime, timezone
import pytz
```

# Create Cities List

```python
cities=501
code=[]
city1=[]
l1=[]
l2=[]

for i in range(0,cities):
    lat=random.uniform(-90,90)
    long=random.uniform(-180,180)
    city = citipy.nearest_city(lat, long)
    code.append(city.country_code)
    city1.append(city.city_name)
    l1.append(lat)
    l2.append(long)
    

df1=pd.DataFrame([code,city1,l1,l2])
df2=df1.T
df2.columns=['Code','City','Lat','Long']
```

# Make API Calls

```python
API='983a6f43b8da565eca1c1d0144337eb9'
```

```python
x1=[]
city=[]
for i in range(0,cities):
    lat2=df2.iloc[i,2]
    long2=df2.iloc[i,3]
    url= f'https://api.openweathermap.org/data/3.0/onecall?lat={lat2}&lon={long2}&units=imperial&appid={API}'
    x1.append(requests.get(url))
    city.append(df2.iloc[i,1])
```

```python
x3=[]
for j in range(0,cities):
    x2=x1[j].json()
    x3.append(x2)
    

df3=pd.DataFrame(x3)
```

```python
city_df=pd.DataFrame(city)
city_df2=city_df.set_axis(['City'], axis=1)
```

```python
location_df=df3[['lat','lon']]
```

```python
df4=df3[['current']]
```

```python
a=[]
for k in range(0,cities):
    b=df4.iloc[k,0]
    a.append(b)
```

```python
df5=pd.DataFrame(a)
```

# Display 501 Call Logs

```python
frames=[city_df2,location_df,df5]
df6=pd.concat(frames,axis=1)
df6
df7=df6.drop(['weather','rain','sunrise','sunset','snow'],axis=1)
display(df7)
```

```python
path='../output_data/weather_data.csv'

df7.to_csv(path)
```

# Regression Analysis by Hemisphere

#     Temperature

```python
df8=df7.query('lat >= 0') #Northern Hemisphere
df9=df7.query('lat < 0')  #Southern Hemishpere
```

```python
x5=df8.lat.values
y5=df8.temp.values
a1,b1 = np.polyfit(x5, y5, 1)
a1,b1
yR5=a1*x5+b1
plt.plot(x5,yR5,color='red')
r = np.corrcoef(x5, y5)
xlab = 'Latitude in Degrees'
ylab = 'Temperature in Fahrenheit'
title = f'Temperature Vs Latitude - Northern Hemisphere\nCorrelation Coeffiecient is = \
{r[0,1]:.4f}'
plt.xlabel(xlab,fontsize=12,fontweight='bold',color='r')
plt.ylabel(ylab,fontsize=12,fontweight='bold',color='r')
plt.title(title,fontsize=20,fontweight='bold',color='g')
plt.scatter(x5,y5)


plt.savefig('../output_data/graph5.png')

print()
```

Temperature has strong linear dependancy on latitude - but to check a residual analysis should be made.

```python
x6=df9.lat.values
y6=df9.temp.values
a2,b2 = np.polyfit(x6, y6, 1)
a2,b2
yR6=a2*x6+b2
plt.plot(x6,yR6,color='red')
r = np.corrcoef(x6, y6)
xlab = 'Latitude in Degrees'
ylab = 'Temperature in Fahrenheit'
title = f'Temperature Vs Latitude - Southern Hemisphere\nCorrelation Coeffiecient is = \
{r[0,1]:.4f}'
plt.xlabel(xlab,fontsize=12,fontweight='bold',color='r')
plt.ylabel(ylab,fontsize=12,fontweight='bold',color='r')
plt.title(title,fontsize=20,fontweight='bold',color='g')
plt.scatter(x6,y6)

plt.savefig('../output_data/graph6.png')
print()
```

Temperature has strong linear dependancy on latitude - but to check a residual analysis should be made.

```python
xlab = 'Latitude in Degrees'
ylab = 'Temperature in Fahrenheit'
title = f'Temperature Vs Latitude - World'
plt.xlabel(xlab,fontsize=12,fontweight='bold',color='r')
plt.ylabel(ylab,fontsize=12,fontweight='bold',color='r')
plt.title(title,fontsize=20,fontweight='bold',color='g')


plt.plot(x5,yR5,color='yellow',label='Northern')
plt.scatter(x5,y5)



plt.plot(x6,yR6,color='purple',label='Southern')
plt.scatter(x6,y6)

plt.legend()

plt.savefig('../output_data/graph7.png')

print()
```

As expected temperature peaks around the equator and we have warm to mild temperatures in the Northern Hemisphere. Northern hemisphere countries are in the warmer seasons, so we would expect the Southern countries to get warmer after some months. 

Looking at the actual data - it shows a curve when we join the hemispheres together.
